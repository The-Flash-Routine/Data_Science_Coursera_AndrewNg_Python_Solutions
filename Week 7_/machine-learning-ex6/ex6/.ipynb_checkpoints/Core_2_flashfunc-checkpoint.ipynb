```python
#for matplotlib
import matplotlib.pyplot as plt
```

```python
#for importing svm module from scikit learn
from sklearn import svm
```

```python
#for loading .mat extension files to program
import scipy.io as sci
```

```python
def plotdata(x,y):
    on=(y==1).ravel()
    off=(y==0).ravel()
    plt.scatter(x[on,0],x[on,1],c='black',marker="+",alpha=0.65,linewidths=0.01)
    plt.scatter(x[off,0],x[off,1],c='yellow',marker='o',alpha=0.65,linewidths=0.01)
    plt.legend(["y=1","y=0"])
```

```python
def plotdecboundary(x,y,clf):
    padding=0.05
    jump=0.02
    xmin=x[:,0].min()-padding
    xmax=x[:,0].max()+padding
    ymin=x[:,1].min()-padding
    ymax=x[:,1].max()+padding
    xx,yy=np.meshgrid(np.arange(xmin,xmax,jump),np.arange(ymin,ymax,jump))
    z=clf.predict(np.c_[xx.ravel(),yy.ravel()])
    z=z.reshape(xx.shape)
    plt.contour(xx,yy,z,alpha=0.8,colors='b',linewidths=0.5)
    
```

```python
#location of data set file
datafileloc="C:/Users/RitKat/Desktop/Projects_Flash/Done using Python_Andrew_Ng_Coursera/Week 7/machine-learning-ex6/ex6/ex6data2.mat"
```

```python
#loading data set from ex6data2.mat file
data=sci.loadmat(datafileloc)
```

```python
print(type(data))
```

```python
#knowing keys present in data
for i in data:
    print(i)
```

```python
#extracting x and y
x=np.array(data["X"])
y=np.array(data["y"])
```

```python
#scatter plotting x1 and x2 wrt y
plotdata(x,y)
```

```python
#training with rbf kernel svm with gamma value 10 and plotting data with decision boundary
clf=svm.SVC(C=1,kernel='rbf',gamma=10)
clf.fit(x,y.ravel())
plotdata(x,y)
plotdecboundary(x,y,clf)
```

```python
#training with rbf kernel svm with gamma value 100 and plotting data with decision boundary
#[improved decision boundary]
clf=svm.SVC(C=1,kernel='rbf',gamma=100)
clf.fit(x,y.ravel())
plotdata(x,y)
plotdecboundary(x,y,clf)
```

```python
#scatter plot of training data
def plottrain(x):
    plt.scatter(x[:,0],x[:,1],marker='x',linewidths=0.01,alpha=0.5)
    plt.xlabel("Latency (ms)")
    plt.ylabel("Throughput (mb/s)")
    plt.xlim(0,30)
    plt.ylim(0,30)
```

```python
#for finding mean and variace columnwise
def findgausparam(x):
    mu=np.mean(x,axis=0)
    vari=np.var(x,axis=0,ddof=1)
    return mu,vari
```

```python
# plot contour of probability
def plotgauss(x,mu,var,xtrain):
    pad=0.1
    xmin=0
    xmax=30
    ymin=0
    ymax=30
    
    xx=np.linspace(xmin,xmax,1000)
    yy=np.linspace(ymin,ymax,1000)
    p=np.zeros((1000,1000))
    for a in range(0,1000):
        for b in range(0,1000):
            pone=calcgauss(float(xx[a]),float(mu[0]),float(var[0]))
            ptwo=calcgauss(float(yy[b]),float(mu[1]),float(var[1]))
            p[a][b]=pone*ptwo
    level=[10**exp for exp in range(-20,0,3)]
    plt.figure(figsize=(5,5))
    
    plt.contour(xx,yy,p,levels=level,colors='black',linestyles='dashed')
    plt.hold(1)
    plottrain(xtrain)
```

```python
#for calculating value of gaussian function 
def calcgauss(x,mu,var):
    hey=float((x-mu)*(x-mu))
    hey=hey/((-2)*var)
    hey=exp(hey)
    hey=hey*(1/sqrt(2*np.pi*var))
    return hey
```

```python
#for selecting best cutoff or threshold value
def selectepi(x,y,mu,var):
    p=calcpval(x,mu,var)
    step=(p.max()-p.min())/1000
    epsilon=np.arange(p.min(),p.max(),step)
    f1opt=-1
    e1opt=-1
    for i in range(0,len(epsilon)):
        pred=(p<epsilon[i]).reshape((-1,1))
        matana=np.hstack((pred,y))
        tp=np.sum(np.logical_and(matana[:,0]==1,matana[:,1]==1))
        fp=np.sum(np.logical_and(matana[:,0]==1,matana[:,1]==0))
        fn=np.sum(np.logical_and(matana[:,0]==0,matana[:,1]==1))
        prec=(tp/(tp+fp))
        recall=(tp/(tp+fn))
        score=(2*prec*recall)/(prec+recall)
        if score>f1opt:
            f1opt=score
            eopt=epsilon[i]
    
    return eopt,f1opt,p
```

```python
def calcpval(x,mu,var):
    p=[];
    m=x.shape[0]
    for i in range(0,m):
        pone=calcgauss(x[i,0],mu[0],var[0])
        ptwo=calcgauss(x[i,1],mu[1],var[1])
        ptot=pone*ptwo
        p.append(ptot)
    p=np.array(p)
    return p
```

```python
def plot_iden_ana(xtrain,mu,var,pval,epsilonsel):
    plotgauss(xtrain,mu,var,xtrain)
    ana=(pval<epsilonsel).ravel()
    plt.scatter(xtrain[ana,0],xtrain[ana,1],edgecolors='r',marker='o',facecolors='none',s=200)
    
```

```python
#function to plot scatter of x1 and x2 wrt y
def plotdata(x,y):
    on=(y==1).ravel()
    off=(y==0).ravel()
    plt.scatter(x[off,0],x[off,1],c='yellow',marker='o')
    plt.scatter(x[on,0],x[on,1],c='black',marker='+')
```

```python
#function to plot decision boundary
def plotdecbound(x,y,clf):
    padding=0.25
    jumping=0.02
    xmin=x[:,0].min()-padding
    xmax=x[:,0].max()+padding
    ymin=x[:,1].min()-padding
    ymax=x[:,1].max()+padding
    xx,yy=np.meshgrid(np.arange(xmin,xmax,jumping),np.arange(ymin,ymax,jumping))
    z=clf.predict(np.c_[xx.ravel(),yy.ravel()])
    z=z.reshape(xx.shape)
    plt.contour(xx, yy, z,colors='b',alpha=0.1)
```

```python
#extracting inpput training examples and output training examples
x=np.array(data["X"])
y=np.array(data["y"])
```

```python
#visualizing data
plotdata(x,y)
```

```python
#ploting x and y scatter and then training with linear svm model with C=1 and ploting decision boundary
#observe effect of changing value of c in subsequent plots
clf=svm.SVC(C=1,kernel='linear')
clf.fit(x,y.ravel())
plotdecbound(x,y,clf)
plotdata(x,y)
```

```python
#ploting x and y scatter and then training with linear svm model with C=100 and ploting decision boundary
#observe effect of changing value of c in subsequent plots
clf=svm.SVC(C=100,kernel='linear')
clf.fit(x,y.ravel())
plotdecbound(x,y,clf)
plotdata(x,y)
```

```python
#ploting x and y scatter and then training with radial basis function kernel(gaussian kernel) svm model with gamma=1 
# and ploting decision boundary
#observe effect of changing value of gamma in subsequent plots
clf=svm.SVC(C=1,kernel='rbf',gamma=1)
clf.fit(x,y.ravel())
plotdecbound(x,y,clf)
plotdata(x,y)
```
